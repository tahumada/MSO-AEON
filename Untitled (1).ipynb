```python
# noirlab

cond = (noirlab_aeon['Instrument'] == 'Goodman') | (noirlab_aeon['Instrument'] == 'TripleSpec4.1')
noirlab_aeon = noirlab_aeon[cond]

cond = noirlab_aeon['Scheduling Constraints'].str.contains('AQ', na=False)
noirlab_aeon = noirlab_aeon[cond]

noirlab_aeon['Propid'] = noirlab_aeon['Propid'].str.replace(r'^N', '2026B-', regex=True)
noirlab_aeon['Propid'] = noirlab_aeon['Propid'].str.replace(r'^25B', '2025B', regex=True)
noirlab_aeon

```

```python
combined_df = pd.concat([noirlab_1, noirlab_2], ignore_index=True)
```

```python
import pandas as pd

# 1. Define the columns you want to pull from noirlab_all 
# (You must include 'number_long' so pandas knows what to match on)
cols_to_pull = [
    'number_long', 'title', 'abstract', 
    'pi_names', 'pi_emails', 'coi_names', 'coi_emails'
]

# 2. Merge the dataframes
noirlab_aeon_updated = pd.merge(
    noirlab_aeon,
    combined_df[cols_to_pull], 
    left_on='Propid', 
    right_on='number_long', 
    how='left'
)

# Optional: Drop the 'number_long' column if you don't want it repeating 
# what's already in the 'Propid' column
noirlab_aeon_updated = noirlab_aeon_updated.drop(columns=['Key','number_long','Sched','Phase','Opt','Acc'])
noirlab_aeon_updated = noirlab_aeon_updated.drop_duplicates()

noirlab_aeon_updated['Sand']
```

```python
# noirlab programs
# 1. Map your existing column names to your desired new names
# Note: I mapped 'pi_names' to 'PI_name', but if you prefer the 'PI' column, 
# just change 'pi_names' to 'PI' in the dictionary below.
column_mapping = {
    'Propid': 'PropID',
    'pi_names': 'PI_name',      
    'pi_emails': 'PI_email',
    'title': 'Title',
    'abstract': 'Abstract',
    'coi_names': 'coIs_name',
    'coi_emails': 'coIs_email',
    'Req': 'time NORMAL'
    # 'Instrument' doesn't need mapping since the name isn't changing
}

# 2. Rename the columns in a new DataFrame
new_df = noirlab_aeon_updated.rename(columns=column_mapping)

# 3. Define the exact final list and order of columns you want
target_columns = [
    'PropID', 'time TIME_CRITICAL', 'time NORMAL', 'Instrument',
    'PI_name', 'PI_email', 'PI_Institution', 'Title', 'Abstract',
    'coIs_name', 'coIs_email'
]

# 4. Reindex to lock in those specific columns
final_noirlab = new_df.reindex(columns=target_columns)
final_noirlab['PI_Institution'] = final_noirlab['PI_email'].str.extract(r'@(.*)\.')
final_noirlab
```
